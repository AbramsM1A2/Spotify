```python
def anadeVertices(G, a): #añade vertices al grafo en funcion de la lista de reproduccion
    G.add_vertices(a)
    return G
def anadeArista(G, a, z):
    for i in range(0,len(a)):
        if(i==len(a)-1):
            G.add_edge(a[-1],a[0],z)
        else:
            G.add_edge(a[i], a[i+1], z)
    return G
```
